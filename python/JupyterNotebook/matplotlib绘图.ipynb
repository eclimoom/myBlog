```python
import numpy as np
from math import sqrt, pi
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['font.sans-serif']=['NSimSun']
plt.title('$y=\sin (x^2)$')
x = np.linspace(-2, 2, 100)
plt.plot(x, np.sin(x ** 2))
plt.show()
```

```python
#用来正常显示中文标签
#仿宋:FangSong,新宋体:NSimSun,楷体:KaiTi,黑体:SimHei,微软雅黑:Microsoft YaHei
plt.rcParams['font.sans-serif']=['NSimSun']
#用来正常显示负号
plt.rcParams['axes.unicode_minus']=False
plt.title('绘图测试')
plt.xlabel("x轴")
plt.ylabel("y轴")
x = np.linspace(-4, 4, 100)
plt.plot(x, x, label='直线')
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.grid(True,color='#EDEDED')
plt.legend()
plt.show()
```

```python
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['font.sans-serif']=['NSimSun']
plt.title('$y=\sin (x^2)$')
# 设置横轴精准刻度
plt.xticks([0, 2, 4, 6, 8, 9, 10],
           ["0m", "2m", "4m", "6m", "8m", "9m", "10m"])
# 设置纵轴精准刻度
plt.yticks([0, 20, 40, 60, 80, 100],
           ["0m", "20m", "40m", "60m", "80m", "100m"])
x=np.linspace(0,10,100)
plt.plot(x,x**2,label='$y=x^2$')
# 设置线形linestyle或ls,'-':实线,'--':短线,':':短点线,'-.':短点相间线
# 设置线宽linewidth或lw参数
plt.plot(x,10*x,label='$y=10x$',alpha=0.4,color='red',ls='--')
plt.legend()
plt.show()
```

```python
# 直方图,是一个特殊的柱状图，又叫做密度图
x=[1,2,2,3,4,4,5]
plt.hist(x,color='green',alpha=0.4)
```

```python
# 条形图：plt.bar()
# 第一个参数是索引。第二个参数是数据值。第三个参数是条形的宽度
x=[1,2,3,4,5]
y=[1,1,3,4,5]
plt.bar(x,y,0.6)
```
